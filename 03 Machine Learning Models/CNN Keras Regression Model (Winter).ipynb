```python
# Import required libraries and read test and train data set
import pandas as pd
import numpy as np
```

## Winter Dataset

```python
winter_dataset = pd.read_csv("Generated Dataset/winter_dataset.csv")
winter_dataset = winter_dataset.drop(["Unnamed: 0"], axis = 1)
# log the GDP and Population
winter_dataset["PopulationSize"] = np.log(winter_dataset["PopulationSize"])
winter_dataset["GDPPerCapita"] = np.log(winter_dataset["GDPPerCapita"])
winter_dataset.head(10)
```

```python
# if the data is too few (less than 3), simply ignore that country
NOC_values_winter = dict(winter_dataset.groupby("CountryCode")["CountryCode"].value_counts())
ignore_value_winter = []
for key in NOC_values_winter:
    if NOC_values_winter[key] <= 3:
        ignore_value_winter.append(key[0])
len(ignore_value_winter)
```

```python
# the data is not enough to split the validation set
# only extract the last year as the test dataset
test_winter = winter_dataset.groupby("CountryCode").apply(lambda x: x.iloc[len(x)-1:])
test_winter = pd.DataFrame(test_winter[(test_winter["TotalParticipants"] != 0) & (~(test_winter["CountryCode"].isin(ignore_value_winter)))]).reset_index(drop = True)
print(test_winter.shape)
NOC_neat_winter = list(test_winter["CountryCode"].unique())
NOC_neat_winter.sort()
train_winter = pd.DataFrame(winter_dataset[(winter_dataset["CountryCode"].isin(NOC_neat_winter)) & (winter_dataset["Year"] != 2016)]).reset_index(drop = True)
train_winter.head(20)
```

```python
train_winter["CountryCode"].unique().shape
```

```python
predictors_winter = ["HostCity","GDPPerCapita","PopulationSize","TotalParticipants", "PrevTotalParticipants", "PrevGoldMedals", "PrevSilverMedals", "PrevBronzeMedals", "PrevTotalMedals",\
                      "PrevSummerTotalParticipants", "PrevSummerGoldMedals", "PrevSummerSilverMedals", "PrevSummerBronzeMedals", "PrevSummerTotalMedals"]
y_train_winter = pd.DataFrame(train_winter["TotalMedals"])
X_train_winter = pd.DataFrame(train_winter[predictors_winter])
y_test_winter = pd.DataFrame(test_winter["TotalMedals"])
X_test_winter = pd.DataFrame(test_winter[predictors_winter])
```

```python
from tensorflow import keras
```

```python
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_squared_error
```

```python
# add CallBacks: early-stopping: stop if there is no improvement on loss more than 100 epochs
earlystopping = keras.callbacks.EarlyStopping(monitor='loss', patience=100, mode='auto')
```

```python
dense_layers = list(range(2, 11))

for dense_layer in dense_layers:
    NAME = "{}-layers-winter".format(dense_layer)
    # always save the best model
    checkpointfunc = keras.callbacks.ModelCheckpoint('Keras Models/{}.model'.format(NAME), monitor='loss', verbose=0,
                                                 save_best_only=True,
                                                 save_weights_only=False, mode='min', period=1)

    model = keras.Sequential()
    model.add(keras.layers.Dense(4, input_dim=14, activation='relu'))
    for i in range(dense_layer):
        model.add(keras.layers.Dense(4, activation='relu'))
    model.add(keras.layers.Dropout(0.3))
    model.add(keras.layers.Dense(1, activation='linear'))
    model.compile(loss='mse', optimizer='adam')
    model.fit(X_train_winter, y_train_winter, epochs=1000, verbose=0, callbacks=[checkpointfunc, earlystopping])
    
    y_test_pred = model.predict(np.array(X_test_winter))
    y_test_pred = list(y_test_pred)
    for i in range(len(y_test_pred)):
        if y_test_pred[i] < 0:
            y_test_pred[i] = 0
    y_test_pred = np.round(y_test_pred).astype("int")
    R2 = explained_variance_score(y_test_winter, y_test_pred)
    MSE = mean_squared_error(y_test_winter, y_test_pred)
    
    print("round", dense_layer, "finish", "R2: ", R2, "MSE: ", MSE)
```

#### 10 dense layers has the highest R2 and the lowest MSE, hence the best model;
### apply the model with 10 dense layers to predict the Winter Dataset

```python
#to load a model
NAME = "{}-layers-winter".format(10)
model = keras.models.load_model('Keras Models/{}.model'.format(NAME))
```

```python
y_test_pred_winter = model.predict(np.array(X_test_winter))
y_test_pred_winter = list(y_test_pred_winter)
for i in range(len(y_test_pred_winter)):
    if y_test_pred_winter[i] < 0:
        y_test_pred_winter[i] = 0
y_test_pred_winter = np.round(y_test_pred_winter).astype("int")
```

```python
y_test_pred_df = pd.DataFrame(y_test_pred_winter)
y_test_df = pd.DataFrame(y_test_winter)
noc_df = pd.DataFrame(NOC_neat_winter)
result_winter = pd.concat([noc_df, y_test_df, y_test_pred_df], axis = 1)
result_winter.columns = ["Country", "True Value", "Predicted Value"]
```
